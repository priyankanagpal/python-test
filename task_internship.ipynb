```python
#Part -1 
#1. Create 4,16,....,52 week moving average(closing price) for each stock and index. This should happen through a function.


#explanation - slow moving averages have larger q, resulting in them responding less to the fluctuations of the stock and being more stable.
#A q-day moving average is, for a series x_t and a point in time t, the average of the past q days: that is, if MA^q_t denotes a moving average process, then:
#MA^q_t = \frac{1}{q} \sum_{i = 0}^{q-1} x_{t - i}

def moving_averages_work(stock,week=[4,16,28,40,52]):
    for values in range(len(week)):
        stock[str(week[values])+"week"]=np.round(stock['Close Price'].rolling(window = week[values]*7, center = False).mean(), 2)
```

```python
#Create rolling window of size 10 on each stock/index. Handle unequal time series due to stock market holidays. 
#You should look to increase your rolling window size to 75 and see how the data looks like. Remember they will
#create stress on your laptop RAM load.
#( Documentation you might need: http://in.mathworks.com/help/econ/rolling-window-estimation-of-state-space-models.html)


def rolling_window(stock,win=[10,75]):
    for values in range(len(win)):
        stock[str(win[values])+"window"]=np.round(stock['Close Price'].rolling(window = win[values] , center = False).mean(), 2)

        
#executing the rolling window function 
rolling_window(tcs)

#checking the columns in the dataset to confirm of the additional window columns being added
tcs.shape
tcs

#plotting the various windows data columns
tcs["10window"].plot(use_index=True)
tcs["75window"].plot(use_index=True)
```

```python
#3. Create the following dummy time series:
   #3.1 Volume shocks - If volume traded is 10% higher/lower 
    #than previous day - make a 0/1 boolean time series for shock,
    #0/1 dummy-coded time series for direction of shock.
# assumption - I considered the previous day to the day for which data was available adding
#forward fill or backward fill could add dates  to the dataset which was not present previously.
# also for first date change is swte to 0 as previous date data is not present
# if change is more than 10% code it 1 else 0

def Volume_shocks_task(st) : 

    st['Total Traded Quantity previously']=tcs["Total Traded Quantity"].shift(1)
    st['changed_volume']= 100*abs(st["Total Traded Quantity"].subtract(st["Total Traded Quantity previously"])).divide(st["Total Traded Quantity previously"])
    st['changed_volume']=np.where(st['changed_volume']>=10,1,0)
```

```python
# Price shocks - If closing price at T vs T+1 has a difference > 2%, 
# then 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.


# assumption - I considered the previous day to the day for which data was available adding
#forward fill or backward fill could add dates  to the dataset which was not present previously.
# also for first date change is set to 0 as previous date data is not present
# if change is more than 10% code it 1 else 0

def price_shocks_task(st) : 

    st['Close Price shifted']=tcs["Close Price"].shift(1)
    st['changed_price']= 100*abs(st['Close Price'].subtract(st['Close Price shifted'])).divide(st['Close Price shifted'])
    st['changed_price']=np.where(st['changed_price']>=2,1,0)

price_shocks_task(tcs)
tcs
```

```python
#3.Pricing black swan - If closing price at T vs T+1 has a difference > 2%,
#then 0/1 boolean time series for shock, 0/1 dummy-coded time series for direction of shock.


def pricing_black_swan(st) : 

    st['Close Price shifted']=tcs["Close Price"].shift(1)
    st['changed_price_swan']= 100*abs(st['Close Price'].subtract(st['Close Price shifted'])).divide(st['Close Price shifted'])
    st['changed_price_swan']=np.where(st['change_priced_swan']>=2,1,0)
```

```python
#4.Pricing shock without volume shock - based on points a & b - Make a 0/1 dummy time series.


def  price_shock_volumne_no_shock(st):
     st['psvn']=np.where(np.logical_and(st['changed_price']==1 , st['changed_volume']==0),1,0)
    
price_shock_volumne_no_shock(tcs)
```

```python
import bokeh
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_file

source = ColumnDataSource(tcs)

p = figure(x_axis_type="datetime", plot_width=800, plot_height=350)
p.line('Date','Close Price',source=source,color='blue')
output_file("st.html")
show(p)
```
